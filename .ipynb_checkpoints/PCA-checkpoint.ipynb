```python
# import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# figures inline, ignore warnings, setting visualisation style
%matplotlib inline 
import warnings 
warnings.filterwarnings("ignore") 
sns.set()

# import algo libraries
from sklearn.utils import resample # for upsampling
from sklearn.preprocessing import StandardScaler # for statndardisation
from sklearn.decomposition import PCA
```

```python
# load datasets
data = pd.read_csv("bank.csv")
```

```python
# basic details about data
data.head()
```

```python
data.info()
```

```python
data.shape
```

```python
# data visualisation
sns.countplot(x=data["subscribed"])
```

```python
# upsampling the lower sample data
data_sub_yes_upsample = resample(data[data["subscribed"]=="yes"], n_samples=2500, random_state=0)
data_upsample = pd.concat([data, data_sub_yes_upsample])
```

```python
sns.countplot(x=data_upsample["subscribed"])
```

```python
# removing the target variable before feature engineering
X = data_upsample.drop('subscribed', axis=1)
y = data_upsample['subscribed']
```

```python
# data encoding
X_encoded = pd.get_dummies(X, dtype=int, drop_first = True )
```

```python
X_encoded
```

```python
scaler = StandardScaler()
X_encoded['age']=scaler.fit_transform(X_encoded[['age']])
X_encoded['balance']=scaler.fit_transform(X_encoded[['balance']])
```

```python
X_encoded
```

```python
# applying PCA on encoded and standard data
pca = PCA()
df_pca = pca.fit_transform(X_encoded)
df_pca = pd.DataFrame(df_pca)
```

```python
df_pca
```

```python
print(pca.explained_variance_ratio_.round(4)[:20]) # individual explained variance
print(pca.explained_variance_ratio_.cumsum().round(2)) # cumulative explained variance
```

```python
#  Create a PCA that will retain 6 features
pca_6 = PCA(n_components=6, whiten=True)

# whiten = True transforms the values of each principal component so that they have zero mean and unit variance.
# if n_components is given 0 to 1, it means the fraction of variation that you want to retain.
# if n_components > 1, it means the number of features you want to retain
```

```python
df_pca_6 = pca_6.fit_transform(X_encoded)
df_pca_6 = pd.DataFrame(df_pca_6)
```

```python
df_pca_6
```

```python
# individual explained variance
print(pca_6.explained_variance_ratio_.round(4))
```

```python
# data visualisation of cumulative variance
variance_cumulative = pca.explained_variance_ratio_.cumsum().round(2)
sns.lineplot(variance_cumulative)
plt.title('Screeplot of Variance Explained %')
plt.xlabel('# of Principal Components')
plt.ylabel('% Variance Explanation')
```

```python
#Getting the Original Features Back
df_orig = pca.inverse_transform(df_pca)
pd.DataFrame(df_orig).round(2)
```
